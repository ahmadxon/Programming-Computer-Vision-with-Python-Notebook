```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread("phone1.jpg", cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("phone2.jpg", cv2.IMREAD_GRAYSCALE)

phone1 = 'phone1.jpg'
phone1 = cv2.imread(phone1, -225)
plt.rcParams["figure.figsize"] = (16,9)
plt.imshow(phone1)
```

```python
phone2 = 'phone2.jpg'
phone2 = cv2.imread(phone2, -225)
plt.rcParams["figure.figsize"] = (16,9)
plt.imshow(phone2)
```

```python
# ORB Detector
orb = cv2.ORB_create()
kp1, des1 = orb.detectAndCompute(img1, None)
kp2, des2 = orb.detectAndCompute(img2, None)
# Brute Force Matching
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des1, des2)
matches = sorted(matches, key = lambda x:x.distance)
matching_result_of_two_phones = cv2.drawMatches(img1, kp1, img2, kp2, matches[:50], None, flags=2)

cv2.imwrite("matching_result_of_two_phones.jpg", matching_result_of_two_phones)
```

```python
matching_result_of_two_phones = 'matching_result_of_two_phones.jpg'
matching_result_of_two_phones = cv2.imread(matching_result_of_two_phones)
plt.rcParams["figure.figsize"] = (16,9)
plt.imshow(matching_result_of_two_phones)
```
